```python
#!/usr/bin/env python3
# Jupyter-friendly full script (no CLI args).
# Loads your JSON (experiment_data.json), computes capacity = packet_count / dur,
# aggregates mean  95% CI (SciPy t-based), and plots ONLY:
#   Capacity vs {loss, delay, rate, reorder} with all *other* parameters FIXED
# (so you see the pure effect of the x-axis parameter).
#
# Edit IN_PATH and FIXED_DEFAULTS as needed.

import json
import re
from pathlib import Path
from typing import Dict, Tuple

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st

# ---------------- Config ----------------
IN_PATH = Path("experiment_data.json")  # <- change if needed
SAVE_CSVS = False                      # True -> write CSVs to ./analysis_out
SAVE_PNGS = False                    # True -> save PNGs to ./analysis_out (in addition to inline)
OUTDIR = Path("analysis_out")

# When plotting a given x-param, hold the *other* params at these fixed values.
# These reflect your example: loss=0%, delay=1 ms, reorder=0%, rate=20 (highest).
FIXED_DEFAULTS = {
    "loss":    {"delay": 1, "reorder": 0, "rate": 20},
    "delay":   {"loss": 0, "reorder": 0, "rate": 20},
    "rate":    {"loss": 0, "delay": 1, "reorder": 0},
    "reorder": {"loss": 0, "delay": 1, "rate": 20},
}

# ------------- Helpers -------------
RUN_RE = re.compile(
    r"^loss_(?P<loss>\d+)_delay_(?P<delay>\d+)_rate_(?P<rate>\d+)_reorder_(?P<reorder>\d+)_(?P<rtype>covert|normal)_#(?P<rep>\d+)$"
)

def parse_run_name(name: str) -> Tuple[int, int, int, int, str, int]:
    m = RUN_RE.match(name)
    if not m:
        raise ValueError(f"Unrecognized run name format: {name}")
    return (
        int(m["loss"]),
        int(m["delay"]),
        int(m["rate"]),
        int(m["reorder"]),
        m["rtype"],
        int(m["rep"]),
    )

def mean_ci95(x: pd.Series) -> Tuple[float, float]:
    """Return (mean, half-width of 95% CI) using t-distribution. NaN if n==0; CI NaN if n==1."""
    x = x.dropna().astype(float)
    n = len(x)
    if n == 0:
        return np.nan, np.nan
    m = x.mean()
    if n == 1:
        return float(m), np.nan
    sem = x.sem(ddof=1)
    tcrit = st.t.ppf(0.975, df=n - 1)
    return float(m), float(tcrit * sem)

# ------------- Load & tidy -------------
def load_json(path: Path) -> Dict:
    with path.open() as f:
        return json.load(f)

def to_long_df(raw: Dict) -> pd.DataFrame:
    rows = []
    for run_name, metrics in raw.items():
        loss, delay, rate, reorder, rtype, rep = parse_run_name(run_name)
        dur = float(metrics["dur"])
        pkt_count = int(metrics["packet_count"])
        cap = pkt_count / dur  # pkts/s
        rows.append(
            dict(
                run_name=run_name,
                loss=loss,
                delay=delay,
                rate=rate,
                reorder=reorder,
                type=rtype,
                rep=rep,
                dur=dur,
                packet_count=pkt_count,
                capacity_pkts_per_s=cap,
            )
        )
    df = pd.DataFrame(rows)
    return df.sort_values(["loss", "delay", "rate", "reorder", "type", "rep"]).reset_index(drop=True)

# ------------- Aggregations (fixed others) -------------
def aggregate_by_param_fixed(df: pd.DataFrame, xparam: str, fixed: dict) -> pd.DataFrame:
    """
    Keep only rows matching `fixed` (for all params except xparam),
    then aggregate capacity vs xparam (covert vs normal), mean  95% CI.
    """
    dff = df.copy()
    # Apply filters for non-x parameters
    for k, v in fixed.items():
        if k == xparam:
            continue
        dff = dff[dff[k] == v]

    if dff.empty:
        raise ValueError(
            f"No rows after filtering with fixed={fixed} for xparam='{xparam}'. "
            f"Adjust FIXED_DEFAULTS to a combination that exists in your data."
        )

    agg = (
        dff.groupby([xparam, "type"])
           .apply(lambda g: pd.Series({
               "n_runs": g.shape[0],
               "capacity_mean": mean_ci95(g["capacity_pkts_per_s"])[0],
               "capacity_ci95": mean_ci95(g["capacity_pkts_per_s"])[1],
           }))
           .reset_index()
           .sort_values(xparam)
    )
    return agg

# ------------- Plotting -------------
def maybe_save(fig: plt.Figure, fname: str):
    if SAVE_PNGS:
        OUTDIR.mkdir(parents=True, exist_ok=True)
        fig.savefig(OUTDIR / fname, dpi=180)

def plot_capacity_vs_param_fixed(df: pd.DataFrame, xparam: str, fixed: dict):
    agg = aggregate_by_param_fixed(df, xparam, fixed)

    # Peek at the table for verification
    try:
        display(agg)
    except NameError:
        print(agg.head())

    fig, ax = plt.subplots(figsize=(7, 4))
    for rtype, sub in agg.groupby("type"):
        x = sub[xparam].values
        y = sub["capacity_mean"].values
        ci = sub["capacity_ci95"].values
        ax.plot(x, y, label=rtype)
        ax.fill_between(x, y - ci, y + ci, alpha=0.2)
        # ax.errorbar(
        #     sub[xparam].values,
        #     sub["capacity_mean"].values,
        #     yerr=sub["capacity_ci95"].values,
        #     marker="o", linestyle="-", linewidth=1.5, markersize=5, label=rtype
        # )
    if xparam == "rate":
        ax.set_xlabel(xparam + " (Mbps)")
    elif xparam == "delay":
        ax.set_xlabel(xparam + " (miliseconds)")
    else:
        ax.set_xlabel(xparam + " (%)")
    ax.set_ylabel("Capacity (pkts/s)")
    ax2 = ax.twinx()
    ax2.set_ylabel("Covert Channel Capacity (Kbps)", labelpad=5)  # label on the right
    ax2.set_ylim(ax.get_ylim())
    ax2.set_yticklabels([f"{(t/1000)}" for t in ax.get_yticks()])
    
    fixed_str = ", ".join(f"{k}={v}" for k, v in fixed.items() if k != xparam)
    # ax.set_title(f"Capacity vs {xparam} (fixed: {fixed_str})")
    ax.legend()
    ax.grid(True, linestyle="--", alpha=0.4)
    plt.show()
    maybe_save(fig, f"capacity_vs_{xparam}__fixed.png")

def plot_all_capacity_fixed(df: pd.DataFrame):
    for xparam in ["loss", "delay", "rate", "reorder"]:
        fixed = FIXED_DEFAULTS.get(xparam, {})
        plot_capacity_vs_param_fixed(df, xparam, fixed)

# ------------- Main -------------
def main():
    if not IN_PATH.exists():
        raise FileNotFoundError(f"JSON file not found: {IN_PATH}")
    raw = load_json(IN_PATH)
    df = to_long_df(raw)

    if SAVE_CSVS:
        OUTDIR.mkdir(parents=True, exist_ok=True)
        df.to_csv(OUTDIR / "runs_long.csv", index=False)

    # Optional: aggregate by exact config (useful for debugging repetitions)
    by_cfg = (
        df.groupby(["loss","delay","rate","reorder","type"])
          .apply(lambda g: pd.Series({
              "n_reps": g.shape[0],
              "capacity_mean": mean_ci95(g["capacity_pkts_per_s"])[0],
              "capacity_ci95": mean_ci95(g["capacity_pkts_per_s"])[1],
          }))
          .reset_index()
    )
    if SAVE_CSVS:
        by_cfg.to_csv(OUTDIR / "by_config_capacity_mean_ci.csv", index=False)

    # Quick peek
    try:
        display(df.head())
        display(by_cfg.head())
    except NameError:
        print(df.head())
        print(by_cfg.head())

    # Plots with other parameters fixed
    plot_all_capacity_fixed(df)

    return df, by_cfg

# Run
df, by_cfg = main()
```

```python
# Load JSON and build long df (assumes your existing to_long_df + mean_ci95 + maybe_save are defined)
# Regex: capture floats/ints for loss, delay, rate, reorder, dup
RUN_RE = re.compile(
    r"^loss_(?P<loss>[0-9.]+)"
    r"_delay_(?P<delay>[0-9.]+)"
    r"_rate_(?P<rate>[0-9.]+)"
    r"_reorder_(?P<reorder>[0-9.]+)"
    r"(?:_dup_(?P<dup>[0-9.]+))?"   # dup is optional
    r"_(?P<rtype>covert|normal)_#(?P<rep>\d+)$"
)

def parse_run_name(name: str) -> Tuple[float, float, float, float, float, str, int]:
    m = RUN_RE.match(name)
    if not m:
        raise ValueError(f"Unrecognized run name format: {name}")
    return (
        float(m["loss"]),
        float(m["delay"]),
        float(m["rate"]),
        float(m["reorder"]),
        float(m["dup"]) if m.group("dup") is not None else 0.0,
        m["rtype"],
        int(m["rep"]),
    )

def to_long_df(raw: Dict) -> pd.DataFrame:
    rows = []
    for run_name, metrics in raw.items():
        loss, delay, rate, reorder, dup, rtype, rep = parse_run_name(run_name)
        dur = float(metrics["dur"])
        pkt_count = int(metrics["packet_count"])
        cap = pkt_count / dur  # pkts/s
        rows.append(
            dict(
                run_name=run_name,
                loss=loss,
                delay=delay,
                rate=rate,
                reorder=reorder,
                dup=dup,
                type=rtype,
                rep=rep,
                dur=dur,
                packet_count=pkt_count,
                capacity_pkts_per_s=cap,
            )
        )
    df = pd.DataFrame(rows)
    return df.sort_values(["loss", "delay", "rate", "reorder", "dup", "type", "rep"]).reset_index(drop=True)


with open("experiment_data_impairment.json") as f:
    raw = json.load(f)
df = to_long_df(raw)

# If your df doesn't include 'dup' (older run_name schema), create it (0 by default)
if "dup" not in df.columns:
    df["dup"] = 0.0

# ---- Define the three regimes (your new realistic settings)
REGIMES = {
    "None":  {"loss": 0.0,   "delay": 0.0,   "rate": 20.0, "reorder": 0.0,  "dup": 0.0},
    "Mid":  {"loss": 0.5,   "delay": 50.0,  "rate": 10.0, "reorder": 0.2,  "dup": 0.001},
    "High": {"loss": 1.0,   "delay": 100.0, "rate": 5.0,  "reorder": 0.4,  "dup": 0.01},
}

# Helper for tolerant equality on floats
def _match(row, params, atol=1e-9, rtol=1e-6):
    checks = []
    for k, v in params.items():
        if k not in row.index:
            continue
        if isinstance(v, float):
            checks.append(np.isclose(float(row[k]), float(v), rtol=rtol, atol=atol))
        else:
            checks.append(row[k] == v)
    return all(checks)

def impairment_regime(row):
    for name, params in REGIMES.items():
        if _match(row, params):
            return name
    return np.nan

# Assign regimes
df_reg = df.copy()
df_reg["regime"] = df_reg.apply(impairment_regime, axis=1)
df_reg = df_reg.dropna(subset=["regime"])

# Aggregate capacity (mean  95% CI) per regime & type
agg = (
    df_reg.groupby(["regime","type"])
          .apply(lambda g: pd.Series({
              "n_runs": g.shape[0],
              "capacity_mean": mean_ci95(g["capacity_pkts_per_s"])[0],
              "capacity_ci95": mean_ci95(g["capacity_pkts_per_s"])[1],
          }))
          .reset_index()
)

display(agg)

# Build nice multi-line x tick labels showing the actual params
def regime_label(name, p):
    return (f"{name}\n"
            f"loss={p['loss']}%, delay={int(p['delay'])}ms\n"
            f"rate={int(p['rate'])}Mbps, reorder={p['reorder']}%, dup={p['dup']}%")

order = ["None","Mid","High"]
xtick_labels = order

# Pivot for plotting (ensure order None -> Mid -> High)
pivot_mean = agg.pivot(index="regime", columns="type", values="capacity_mean").reindex(order)
pivot_ci   = agg.pivot(index="regime", columns="type", values="capacity_ci95").reindex(order)

# --- Plot: Bar chart covert vs normal with params on x-axis ---
fig, ax = plt.subplots(figsize=(12, 5))
x = np.arange(len(order))
width = 0.35

for i, t in enumerate(pivot_mean.columns):
    y    = pivot_mean[t].values
    yerr = pivot_ci[t].values
    ax.bar(x + (i - 0.5) * width, y, width=width, yerr=yerr, capsize=4, label=t)

ax.set_xticks(x)
ax.set_xticklabels(xtick_labels)
ax.set_xlabel("Impairment Regime")
ax.set_ylabel("Capacity (pkts/s)")

# Right-side y-axis: mirror ticks, show as Kbps for covert bit-per-packet intuition
ax2 = ax.twinx()
ax2.set_ylim(ax.get_ylim())
yt = ax.get_yticks()
ax2.set_yticks(yt)
ax2.set_ylabel("Covert Channel Capacity (Kbps)", labelpad=6)
ax2.set_yticklabels([f"{t/1000:.1f}" for t in yt])

ax.legend(title="Channel Type")
ax.grid(True, axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
maybe_save(fig, "impairment_regimes_realistic.png")
plt.show()
```

```python
# --- Channel capacity & eBPF overhead under Low / Med / High impairment regimes ---

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re, json
from pathlib import Path

# Thresholds for binning
LOSS_LOW,  DELAY_LOW,  REORDER_LOW,  RATE_LOW  = 0,  1,  0, 20
LOSS_MED,  DELAY_MED,  REORDER_MED,  RATE_MED  = 10, 21, 5, 12
LOSS_HIGH, DELAY_HIGH, REORDER_HIGH, RATE_HIGH = 20, 61, 10, 4

def impairment_regime(row):
    low  = (row["loss"] == LOSS_LOW  and row["delay"] == DELAY_LOW  and row["reorder"] == REORDER_LOW  and row["rate"] == RATE_LOW)
    med  = (row["loss"] == LOSS_MED  and row["delay"] == DELAY_MED  and row["reorder"] == REORDER_MED  and row["rate"] == RATE_MED)
    high = (row["loss"] == LOSS_HIGH and row["delay"] == DELAY_HIGH and row["reorder"] == REORDER_HIGH and row["rate"] == RATE_HIGH)
    if low:  return "Low"
    if med:  return "Med"
    if high: return "High"
    return np.nan

# If avg_ns_per_packet is missing, derive it from ebpf_stats in the JSON once
if "avg_ns_per_packet" not in df.columns:
    AVG_NS_RE = re.compile(r"Average time per packet:\s*([0-9.]+)\s*ns", re.IGNORECASE)
    with open(IN_PATH) as f:
        raw = json.load(f)
    avg_ns_map = {}
    for run_name, metrics in raw.items():
        txt = metrics.get("ebpf_stats", "")
        m = AVG_NS_RE.search(txt) if isinstance(txt, str) else None
        avg_ns_map[run_name] = float(m.group(1)) if m else np.nan
    df["avg_ns_per_packet"] = df["run_name"].map(avg_ns_map)

# Build regimes
df_reg = df.copy()
df_reg["regime"] = df_reg.apply(impairment_regime, axis=1)
df_reg = df_reg.dropna(subset=["regime"])

# --- Aggregate capacity (mean  95% CI) per regime & type
agg_cap = (
    df_reg.groupby(["regime","type"])
          .apply(lambda g: pd.Series({
              "n_runs": g.shape[0],
              "capacity_mean": mean_ci95(g["capacity_pkts_per_s"])[0],
              "capacity_ci95": mean_ci95(g["capacity_pkts_per_s"])[1],
          }))
          .reset_index()
)

# --- Aggregate eBPF overhead (mean  95% CI) per regime & type
agg_over = (
    df_reg.groupby(["regime","type"])
          .apply(lambda g: pd.Series({
              "n_runs": g.shape[0],
              "overhead_mean": mean_ci95(g["avg_ns_per_packet"])[0],
              "overhead_ci95": mean_ci95(g["avg_ns_per_packet"])[1],
          }))
          .reset_index()
)

display(agg_cap)
display(agg_over)

# Pivot for plotting (ensure order Low -> Med -> High)
order = ["Low","Med","High"]
p_cap_mean = agg_cap.pivot(index="regime", columns="type", values="capacity_mean").reindex(order)
p_cap_ci   = agg_cap.pivot(index="regime", columns="type", values="capacity_ci95").reindex(order)

p_ov_mean  = agg_over.pivot(index="regime", columns="type", values="overhead_mean").reindex(order)
p_ov_ci    = agg_over.pivot(index="regime", columns="type", values="overhead_ci95").reindex(order)

# --- Plot: two stacked panels (top = capacity, bottom = eBPF overhead)
fig, (ax, axb) = plt.subplots(2, 1, figsize=(7, 7), sharex=True)
x = np.arange(len(order))
width = 0.35

# Capacity bars
for i, t in enumerate(p_cap_mean.columns):
    y    = p_cap_mean[t].values
    yerr = p_cap_ci[t].values
    ax.bar(x + (i - 0.5) * width, y, width=width, yerr=yerr, capsize=4, label=t)
ax.set_ylabel("Capacity (pkts/s)")
ax.set_xticks(x)
ax.set_xticklabels(order)
ax.legend(title="Channel Type")
ax.grid(True, axis="y", linestyle="--", alpha=0.4)

# (Optional) right-side axis mirroring capacity in k pkts/s
ax2 = ax.twinx()
ax2.set_ylim(ax.get_ylim())
yt = ax.get_yticks()
ax2.set_yticks(yt)
ax2.set_ylabel("Capacity (k pkts/s)", labelpad=6)
ax2.set_yticklabels([f"{t/1000:.1f}" for t in yt])

# Overhead bars
for i, t in enumerate(p_ov_mean.columns):
    y    = p_ov_mean[t].values
    yerr = p_ov_ci[t].values
    axb.bar(x + (i - 0.5) * width, y, width=width, yerr=yerr, capsize=4, label=t)
axb.set_xlabel("Impairment Regime")
axb.set_ylabel("Avg BPF time (ns/packet)")
axb.set_xticks(x)
axb.set_xticklabels(order)
axb.grid(True, axis="y", linestyle="--", alpha=0.4)

plt.tight_layout()
maybe_save(fig, "impairment_regimes_with_bpf_overhead.png")
plt.show()
```
